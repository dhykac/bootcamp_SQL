### Question 1 Point 40 : Find the 3rd Highest Salary

You will be given a table. 

The table has 2 columns: `employee`, and `salary`.

Please output the 3rd Highest Salary! 


#### Example 1:
| employee | salary |
| --- | --- |
| Abdul | 100 |
| Budi| 200 |
| Charlie | 100 |
| Dimas | 500 |

#### Answer 1: 
`100`

Yup. Just output the 3rd highest salary. Please, do NOT give any Employee Name, do NOT give duplicate values. We know that 100 is the 3rd Highest Salary, and there are 2 Employees whose salary is 100. But we just care about the VALUE of the 3rd highest salary. Once again, no need to write duplicate values, no need to write the employee name

#### Untuk Class Manager

Jika ada:
- Duplicate row (muncul lebih dari hanya 1 baris saja)
- Nama ikutan diberikan dalam jawaban
- Kolom lain ikut diberikan dalam jawaban

Maka hal tersebut merupakan sebuah kesalahan

```python
employee_table = pd.DataFrame({
    'employee':['Alice', 'Benn', 'Charles', 'Dorothy', 'Emma', 'Franklin'],
    'salary':[11,43,33,55,22,33]
})
```

```python
### silakan dilengkapi

query = """
SELECT salary
FROM ((SELECT * from employee_table
       ORDER BY salary DESC
       LIMIT 3) as third_highest)
ORDER BY third_highest.salary ASC 
LIMIT 1
"""

sql_run(query)
```

```python
item_bought = pd.DataFrame({
    'buyer':['A', 'B', 'C', 'D', 'E', 'F'],
    'item':['Asus', 'Lenovo', 'Lenovo', 'Acer', 'Acer', 'Acer']
})
```

```python
## silakan dilengkapi

query = """ 
WITH base_table AS
(SELECT item, count(*) quantity
FROM item_bought
GROUP BY item)

SELECT item
FROM base_table
WHERE quantity != (SELECT MAX(quantity) FROM base_table)
AND quantity != (SELECT MIN(quantity) FROM base_table)
"""
sql_run(query)
```
